```python
import numpy as np
import matplotlib.pyplot as plt

def calculate_and_plot_probabilities(target_step):
    # Initialize an array to store the probabilities of reaching each step, with step 1 having a probability of 1
    probs = np.zeros(target_step + 1)
    probs[1] = 1  # Starting on step 1

    # Calculate probabilities for steps 2 through target_step
    for current_step in range(2, target_step + 1):
        # You can reach the current step either from the previous step or the step before that
        probs[current_step] = 0.5 * probs[current_step - 1]
        if current_step > 2:
            probs[current_step] += 0.5 * probs[current_step - 2]

    # Plotting
    plt.figure(figsize=(12, 8))
    plt.plot(range(1, target_step + 1), probs[1:], marker='o', markersize=3, linestyle='-', linewidth=1)
    plt.title(f'Probability of Landing on Each Step up to {target_step}')
    plt.xlabel('Step Number')
    plt.ylabel('Probability')
    plt.xticks(range(0, target_step + 1, max(1, target_step // 10)))  # Set x-axis ticks to show every 10th step for clarity, adjusting for small numbers
    plt.grid(True)
    plt.tight_layout()
    plt.show()


calculate_and_plot_probabilities(100)
```
